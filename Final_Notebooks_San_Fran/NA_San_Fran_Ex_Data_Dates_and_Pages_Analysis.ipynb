```python
import matplotlib.pyplot as plt
```

```python
# Source of Code for Pie Charts in Google Colab: https://colab.research.google.com/drive/11Gd3rmm7TR-ugw4blV6RRko-zf-d6fmF

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
x = [73, 56, 47, 34, 34, 40, 26]
labels = ['19/04/1906', '20/04/1906', '21/04/1906', '22/04/1906', '23/04/1906', '24/04/1906', '25/04/1906']

fig, ax = plt.subplots()
ax.pie(x, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
ax.set_title('Dates')
plt.tight_layout()
```

```python
# Source of Code for Bar Graphs in Google Colab: https://colab.research.google.com/notebooks/charts.ipynb
# Source of Code for Bar Graph Formatting in Google Colab: https://colab.research.google.com/github/Tanu-N-Prabhu/Python/blob/master/Mastering_the_Bar_Plot_in_Python.ipynb

# Bar graph for dates:
pub_date_amounts = [73, 56, 47, 34, 34, 40, 26]
pub_date_labels = ['19/04', '20/04', '21/04', '22/04', '23/04', '24/04', '25/04']

# Passing the parameters to the bar function, this is the main function which creates the bar plot
plt.bar(pub_date_labels, pub_date_amounts)

# This is the location for the annotated text
i = 1.0
j = 1
# Annotating the bar plot with the values (total count)
for i in range(len(pub_date_labels)):
    plt.annotate(pub_date_amounts[i], (-0.1 + i, pub_date_amounts[i] + j))

# Namimg the x and y axis
plt.xlabel('Dates')
plt.ylabel('Articles in Corpus')

# Creating the legend of the bars in the plot
plt.legend(labels = ['Total Articles'])
# Giving the tilte for the plot
plt.title("Article Distribution by Date in Corpus")

plt.show()
```
