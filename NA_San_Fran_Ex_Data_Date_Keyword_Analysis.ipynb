```python
#Data organizing and filtering method from NLP class
#Import the dataset "San Fran EExtended Sample"
import pandas as pd

df = pd.read_excel('/content/NLP_Winter_2024/NA_San_Fran_Extended_Sample.xlsx')

#Display DataFrame 'df'
df.head()
```

```python
#What is the Length of the DataFrame?

print(f"The DataFrame has {df.shape[0]} rows.")
```

```python
#How many different newspaper titles are tthere?
num_unique_titles = df['Full_Name'].nunique()

print(f"There are {num_unique_titles} different paper titles in the dataset.")
```

```python
#Filtering Dates
#Filter the DataFrame to include only dates of 19/04/1906.
df_date_19 = df[df['Day'].astype('str').str.contains('19/04/1906', na = False)]

#Display the filtered DataFrame
df_date_19.head()
```

```python
# How many rows are these?
print(f"This DataFrame for 19/04/1906 has {df_date_19.shape[0]} rows.")
```

```python
# Only date of 20/04/1906
df_date_20 = df[df['Day'].astype('str').str.contains('20/04/1906', na = False)]

# How many rows are these?
print(f"This DataFrame for 20/04/1906 has {df_date_20.shape[0]} rows.")

# Only date of 21/04/1906
df_date_21 = df[df['Day'].astype('str').str.contains('21/04/1906', na = False)]

# How many rows are these?
print(f"This DataFrame for 21/04/1906 has {df_date_21.shape[0]} rows.")

# Only date of 22/04/1906
df_date_22 = df[df['Day'].astype('str').str.contains('22/04/1906', na = False)]

# How many rows are these?
print(f"This DataFrame for 22/04/1906 has {df_date_22.shape[0]} rows.")

# Only date of 23/04/1906
df_date_23 = df[df['Day'].astype('str').str.contains('23/04/1906', na = False)]

# How many rows are these?
print(f"This DataFrame for 23/04/1906 has {df_date_23.shape[0]} rows.")

# Only date of 24/04/1906
df_date_24 = df[df['Day'].astype('str').str.contains('24/04/1906', na = False)]

# How many rows are these?
print(f"This DataFrame for 24/04/1906 has {df_date_24.shape[0]} rows.")

# Only date of 25/04/1906
df_date_25 = df[df['Day'].astype('str').str.contains('25/04/1906', na = False)]

# How many rows are these?
print(f"This DataFrame for 25/04/1906 has {df_date_25.shape[0]} rows.")
```
